```python
import pandas as pd
import plotly.express as px

# loads the data
df = pd.read_csv("cleaned_rental_data (1).csv")
# df = pd.read_csv("cleaned_rental_data.csv")

# filters for numeric columns
numeric_df = df.select_dtypes(include=['number'])

# computes the correlation matrix
correlation_matrix = numeric_df.corr().round(2)

# resets index to be read properly
corr_df = correlation_matrix.reset_index().melt(id_vars="index")
corr_df.columns = ['Variable 1', 'Variable 2', 'Correlation']

# plots the interactive heatmap
fig = px.imshow(
    correlation_matrix,
    text_auto=True,
    color_continuous_scale="RdBu_r",
    zmin=-1,
    zmax=1,
    labels=dict(color="Correlation"),
    title="Correlation Heatmap of Boston Rental Data"
)

fig.update_layout(
    width=900,
    height=800,
    xaxis_title="",
    yaxis_title="",
    xaxis_tickangle=45
)

fig.show()
```

```python
# Converts the chart to html format and creates the file
fig.write_html("corr_heatmap.html")

from google.colab import files
files.download('corr_heatmap.html')
```
